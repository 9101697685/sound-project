To facilitate analysis and comparison of two flow meters (ultrasonic and orifice), it is important to ensure that they have the same unit of measurement. In this case, we need to convert the unit of the orifice flow meter from Tons per Hour (t/h) to the unit used by the ultrasonic meter, which is Cubic Meters per Hour (m³/h). By doing so, we can establish a common basis for evaluating and comparing their flow rate readings.

### How to Convert the Flow Rate of an Orifice Sensor (96KJ04) from Tons per Hour (t/h) to Cubic Meters per Hour (m³/h)?


Given (96KJ04 orifice flow meter parameters):
- Temperature = 165°C
- Pressure = 5 Bar ABS

To perform the conversion, follow these steps:

1. Retrieve the Specific Value:
   - Input the temperature and pressure values into the TLV website to obtain the specific value.
   - In this case, the specific value is determined as 0.3889 m^3/kg or 388.9 m^3/t.

2. Calculate the Conversion:
   - For example, if the steam flow rate is 20 tons per hour, multiply it by the specific value to convert to cubic meters per hour.
   - The calculation would be as follows: 20 t/h * 388.9 m^3/t = 7778 m^3/h.

TLV website:
https://www.tlv.com/global/TI/calculator/steam-flow-rate-through-valve.html

### Create a function that plots volumetric flow rate data for visualization to ensure the data is clear and reliable

```python
def plot_velumetric_flow_data(filename):
    
    def clean_flexim_data(df):
        # convert non-numeric values to nan value
        df.iloc[0] = np.nan
        df.iloc[1] = np.nan

        # delete rows with nan values
        df = df.dropna()

        # Reset index after drop
        df = df.reset_index(drop=True)
        return df
    
    # Read data column from the flexim excel file
    df = pd.read_excel(filename, usecols='B')
    # Clean the data based on the clean function
    df = clean_flexim_data(df)

    # Convert the column containing date and time to datetime type
    df['DateTime'] = pd.to_datetime(df['Unnamed: 1'])

    # Extract the date and time into separate columns
    df['Date'] = df['DateTime'].dt.date
    df['Time'] = df['DateTime'].dt.time

    # Read velumetric flow rate column
    velumetric_flow_rate = pd.read_excel(filename, usecols='D')
    # Clean the velumetric flow rate data
    velumetric_flow_rate = clean_flexim_data(velumetric_flow_rate)

    # Convert the 'Time' values to strings
    time_strings = [str(time) for time in df['Time']]

    # Set the figure size
    plt.figure(figsize=(30, 5))

    # Plot the desired column against the time strings
    plt.plot(time_strings, velumetric_flow_rate, color='blue', marker='s', linestyle='-', markersize=6)

    # Add labels and title
    plt.xlabel("Time", fontsize=20)
    plt.ylabel("A: volumetric flow rate [$m^3$/h]", fontsize=18)
    plt.title("Steam Flow", fontsize=20)

    # Rotate x-axis tick labels by 45 degrees
    plt.xticks(rotation=45)

    plt.grid()
    plt.show()

plot_velumetric_flow_data('fake_ultrasonic_data.xlsx')
```

### Create a function that calculates the mean, median, and standard deviation for both orifice and ultrasonic data, allowing for a comparison of their results

```python
def calculate_flow_meter_statistics(orifice_filename, ultrasonic_filename):
    
    def clean_flexim_data(df):
        # convert non-numeric values to nan value
        df.iloc[0] = np.nan
        df.iloc[1] = np.nan

        # delete rows with nan values
        df = df.dropna()

        # Reset index after drop
        df = df.reset_index(drop=True)
        return df
    
    # Ignore warnings
    warnings.filterwarnings("ignore")

    # Represent the data of orifice sensor 96KJ04
    df_orifice = pd.read_excel(orifice_filename, usecols='AO') 
    # Convert orifice flow meter from t/h to m^3/h
    specific_value = 388.9  # m³/t
    df_orifice = df_orifice * specific_value
    orifice_mean = np.mean(df_orifice).item()
    orifice_median = np.median(df_orifice).item()
    orifice_std = np.std(df_orifice).item()

    # Represent the data of ultrasonic sensor (the flow velocity is given as 1.21 m/s)
    # Read flow velocity from the excel file
    df_ultrasonic = pd.read_excel(ultrasonic_filename, usecols='D')
    # Clean the data
    df_ultrasonic = clean_flexim_data(df_ultrasonic)
    ultrasonic_mean = np.mean(df_ultrasonic).item()
    ultrasonic_median = np.median(df_ultrasonic).item()
    ultrasonic_std = np.std(df_ultrasonic).item()

    # Format the numbers with 2 decimal places
    orifice_mean = format(orifice_mean, '.2f')
    orifice_median = format(orifice_median, '.2f')
    orifice_std = format(orifice_std, '.2f')
    ultrasonic_mean = format(ultrasonic_mean, '.2f')
    ultrasonic_median = format(ultrasonic_median, '.2f')
    ultrasonic_std = format(ultrasonic_std, '.2f')

    # Create a table to compare the results
    table_data = [
        ["", "Mean", "Median", "Standard Deviation"],
        ["Orifice Plate Flow Meter", orifice_mean, orifice_median, orifice_std],
        ["Ultrasonic Flow Meter", ultrasonic_mean, ultrasonic_median, ultrasonic_std]
    ]
    table = tabulate(table_data, headers="firstrow")

    # Print the table
    print(table)

    # Plot the table as a bar chart
    x_labels = ["Mean", "Median", "Standard Deviation"]
    orifice_values = [float(orifice_mean), float(orifice_median), float(orifice_std)]
    ultrasonic_values = [float(ultrasonic_mean), float(ultrasonic_median), float(ultrasonic_std)]

    plt.bar(x_labels, orifice_values, label="Orifice Plate Flow Meter")
    plt.bar(x_labels, ultrasonic_values, label="Ultrasonic Flow Meter")
    plt.xlabel("Statistics")
    plt.ylabel("Values")
    plt.title("Flow Meter Statistics")
    plt.legend()
    plt.show()

calculate_flow_meter_statistics('real_orifice_data.xlsx', 'fake_ultrasonic_data.xlsx')
```

### Create a function that compares the flow measurements from both orifice and ultrasonic flowmeters and visualizes the results on a graph

```python
def create_flow_meter_comparison_plot(orifice_filename, ultrasonic_filename):
    
    def clean_flexim_data(df):
        # convert non-numeric values to nan value
        df.iloc[0] = np.nan
        df.iloc[1] = np.nan

        # delete rows with nan values
        df = df.dropna()

        # Reset index after drop
        df = df.reset_index(drop=True)
        return df

    # Read the orifice sensor data 96KJ04
    df_orifice = pd.read_excel(orifice_filename, usecols='AO')
    
    # Convert orifice flow meter from t/h to m^3/h
    specific_value = 388.9  # m³/t
    df_orifice = df_orifice * specific_value

    # Read the ultrasonic data
    df_ultrasonic = pd.read_excel(ultrasonic_filename, usecols='D')

    # Both datasets must have the same data size to plot, so find the one with the smaller size first:
    # Determine the smaller size between the two data sets
    min_size = min(len(df_orifice), len(df_ultrasonic))
    # Then, adjust the size of both data sets
    df_orifice = df_orifice.iloc[:min_size, :]
    df_ultrasonic = df_ultrasonic.iloc[:min_size, :]

    # Clean the ultrasonic data
    df_ultrasonic = clean_flexim_data(df_ultrasonic)

    # Create plot for orifice data (blue)
    plt.plot(df_orifice, color='blue', label='Orifice flowmeter')

    # Create plot for ultrasonic data (red)
    plt.plot(df_ultrasonic, color='red', label='Ultrasonic flowmeter')

    # Add labels and title
    plt.xlabel("The number of data points")
    plt.ylabel("The steam flow rate")
    plt.title("Comparison of Flow Meters")

    plt.legend()
    plt.show()

create_flow_meter_comparison_plot('real_orifice_data.xlsx', 'fake_ultrasonic_data.xlsx') 
```
